```python
import pandas as pd
import numpy as np
import json
import copy
import matplotlib.pyplot as plt
```

```python
with open('glasgow_house_prices_0.json','r') as handle:
    house_data = json.load(handle)
```

```python
data = {'dates':[],'address':[],'prices':[]}
for k,v in house_data.items():
    for kk,vv in v.items():
        if kk in data:
            data[kk]+=vv
            
        
```

```python
df = pd.DataFrame.from_dict(data)
dates = df.dates
df.prices = df.prices.astype(int)
df.shape
```

```python
help(pd.to_datetime)
```

```python
df.dates=pd.to_datetime(dates,errors='coerce',infer_datetime_format=True,utc=True)
```

```python
df.dates.min()
```

```python
df = df[df.prices.between(1000,800000)] 
df['year'] = np.zeros(df.shape[0]).astype(int)
df.prices.values[[df.prices.unique().argmin()]]
```

```python
dates = [[f'200{yr-1}-1-1',f'200{yr}-1-1' ]if yr<10 else [f'20{yr}-1-1',f'20{yr+1}-1-1']
         for yr in range(1,23)]
dates
```

```python
for yr_n,yrs in enumerate(dates):
    print(yrs)
    mask =(df['dates']>yrs[0])&(df['dates'] <= yrs[1])
    df.loc[mask,'year']=int(2000+yr_n+1)
df = df[df['year']!=0]
```

```python
df.plot(x='dates',y='prices',kind='scatter')
```
